```python
n = int(input(" Enter The number of students:")) 

stud = [] 

for i in range(0, n): 

    stud_name = input('Enter the name of student: ') 

    print(stud_name) 
 

    stud_rollno = input('Enter the roll number of student: ')

    print(stud_rollno) 
 

    sub1 = eval(input('Enter the marks in CS601: ')) 

    print(sub1) 
 

    sub2 = eval(input('Enter the marks in CS602B: ')) 

    print(sub2) 
 

    sub3 = eval(input('Enter the marks in  CS603: '))

    print(sub3) 

 
    tot = (sub1 + sub2 + sub3) 

    print("Total is: ", tot) 

 
    avg = tot / 3 

    print("Average is :", avg) 

    stud.append({ 

                            'Name': stud_name, 

                            'Rollno': stud_rollno, 

                            'Sub1': sub1, 

                            'Sub2': sub2, 

                            'Sub3': sub3, 

                            'Total': tot, 

                            'Average': avg 

                            }) 
 

for student in stud: 

    print('\n') 

    for key, value in list(student.items()): 

        print('{0}: {1}'.format(key, value)) 
```

```python

def change_date(convert_date):
    changed_date_list = []

    months = {
        "JAN": 1,
        "FEB": 2,
        "MAR": 3,
        "APR": 4,
        "MAY": 5,
        "JUN": 6,
        "JUL": 7,
        "AUG": 8,
        "SEP": 9,
        "OCT": 10,
        "NOV": 11,
        "DEC": 12
    }
    input_split = convert_date.split('-')

    #year:
    if int(input_split[2])>=0:
        year = '20' + input_split[2]
    else:
        year = '19' + input_split[2]

    changed_date_list.append(year)

    #month:
    month = months[input_split[1].upper()]
    changed_date_list.append(month)

    #day:
    day = input_split[0]
    changed_date_list.append(day)

    changed_date_list = [str(item) for item in changed_date_list]
    return changed_date_list


def main():
    user_date = input("Enter a date in this format: dd-MMM-yy \n") 
    print(','.join((change_date(user_date))))

main()
```
