```python
import random

count = 0
jumb_word = ""
punct = ("!","?","|",",",".")

inputfile = input("What is The input file name?")

with open(inputfile, 'r') as fin:
	for line in fin.readlines(): #  To Read Input  line by line in .txt file

		for word in line.split(): # Read word by word in each line

			if len(word) > 3: # Divided in the Case If word length is Greater Than 3

				''' CASE 1 :If word ends with punctuation
					Remove firstletter, lastletter and punctuation
					Shuffle the words
					Add the removed letters first letter,last letter
					Add the removed letters (punctuation mark)'''

				if word.endswith(punct):
					letter1 = word[1:-2]
					letter1 = random.sample(letter1, len(letter1))
					letter1.insert(0, word[0])
					letter1.append(word[-2])
					letter1.append(word[-1])

					'''If there is no punctuation mark
					Remove first letter and last letter
                    and Do The Same AS In Case 1
					Append the word and " " to the previous words'''

				else:
					letter1 = word[1:-1]
					letter1 = random.sample(letter1, len(letter1))
					letter1.insert(0, word[0])
					letter1.append(word[-1])
					jumb_word = jumb_word + ''.join(letter1) + " "

			'''If word length <3
				just append the word and add " " to the previous words'''

		else:
			jumb_word = jumb_word + word + " "


with open(("Scrambled.txt"), 'a+') as fout:
	fout.write(jumb_word + "\n")

jumb_word = ""
```
