```python
list1 = [[4, 5, 6], [1, 2, 3], [10, 11, 12], [7, 8, 9]]

ele = 0
MAX_ind = 0
sum_max = 0
for list in list1:
    sum_list = 0
    for i in list:
        sum_list += i
    if sum_list > sum_max:
        sum_max = sum_list
        MAX_ind = ele
    ele += 1

print(list1[MAX_ind])
```

```python
givlst = [1,2,3,4] 
searchlst = [2,3] 


def sublist(searchlst, givlst):
   
    ind1 = 0  
    ind2 = 0 

    # go through indexes of searchlst
    while ind1 < len(searchlst):

        # search for an item in givlst matching item in searchlst at index ind1
        while ind2 < len(givlst) and searchlst[ind1] != givlst[ind2]:
            ind2 += 1

        # if we run out of items in givlst then searchlst is not a sublist of givlst
        if ind2 == len(givlst):
            return False
        ind1 += 1

    # if every item in searchlst has been matched to an item in givlst, from left to right
    return True

print( sublist(searchlst, givlst))
```
