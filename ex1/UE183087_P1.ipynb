```python
# 1.Python program to print Even Numbers in given range
print("1.Python program to print Even Numbers in given range")
start, end = 50, 80


for num in range(start, end + 1):
	

	if num % 2 == 0:
		print(num, end = " ")
print("\n")
  
# 2.Python program to print  Sum of N natural Numbers in given range

print("\n2.Python program to print  Sum of N natural Numbers in given range")

num = int(input("Enter a number: "))  
if num < 0:  
   print("Enter a positive number")  
else:  
   sum = 0  
   #  while loop to iterate  till zero  
   while(num > 0):  
       sum += num  
       num -= 1  
   print("The sum is",sum)  

# 3.Program to check if a number is prime or not
print("\n3.Program to check if a number is prime or not")
num =int(input("Enter The Number:"))

# define a flag variable
flag = False

# prime numbers are greater than 1
if num > 1:
    # check for factors
    for i in range(2, num):
        if (num % i) == 0:
            # if condition is True, set flag to True
            flag = True
            # break out of loop
            break

# check if flag is True
if flag:
    print(num, "is not a prime number")
else:
    print(num, "is a prime number")
    
# 4.Program to display the Fibonacci sequence up to n-th term
print("\n4.Program to display the Fibonacci sequence up to n-th term")
upton = int(input("Enter The Range"))

# first two terms
n1, n2 = 0, 1
count = 0

# check if the number of terms is valid
if upton <= 0:
   print("Please enter a positive integer")
elif upton == 1:
   print("Fibonacci sequence upto",upton,":")
   print(n1)
else:
   print("Fibonacci sequence is:")
   while count < upton:
       print(n1)
       nth = n1 + n2
       # update values
       n1 = n2
       n2 = nth
       count += 1
```
